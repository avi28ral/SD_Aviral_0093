```python
menu = {
    "Cold Tea":90,
    "Chowmein":70,
    "Pizza":250,
    "Burger":150
}
print("Welcome to our Res.......")
print(menu)
order_amount=0
while True :
  order = input("Enter Your Order").title()
  if order in menu:
    order_amount += menu[order]

  else:
    print("item not available in menu")
  choice = input("wish to add another item? (Yes/No)").title()
  if choice != "Yes" :
    break

print(f"your total bill is {order_amount}")
```
