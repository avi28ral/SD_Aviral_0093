```python
def count_test(n):
  count = 1
  while count <= n:
    yield count
    count = count+1
```

```python
c = count_test(5)
```

```python
print(c)
```

```python
type(c)
```

```python
for i in c:
  print(i)
```

```python
def count(x):
  c = 0
  if x == 0:
    return 1
  while x != 0 :
      c += 1
      x = x//10

  return c
```

```python
print(count(1234))
```

**POLYMORPHISM**

```python
class UPI:
  def pay(self):
    print("Payment through UPI")
class Card:
  def pay(self):
    print("Payment through Card")
for payment in [UPI(), Card()]:
  payment.pay()
```
