```python
expenses ={}
choice = input("wish to add a category(yes/no)").lower()
while(choice == "yes"):
  category = input("add a category").lower()
  cost = int(input("enter expense in this category"))
  expenses[category] = cost
  choice = input("wish to add a category(yes/no)").lower()
wish = input("want to update expenses in a category? (yes/no) ").lower()
while(wish == "yes"):
  a = input("enter category to update cost").lower()
  if a in expenses.keys():
    updation = int(input("enter amount to be added"))
    expenses[a] += updation
  else:
    print("invalid category")
  wish = input("want to update expenses in a category? (yes/no) ").lower()
```
